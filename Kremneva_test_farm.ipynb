```python
class Farmer:
  def __init__(self, name):
    self.name = name
    self.money = 300000
    self.farm = Farm()

  def __str__(self):
    return f'фермер {self.name}'

  def buy_animal(self): # покупаем животных
    animal_type = input('Введите животное, которое хотите купить - корова, коза, курица, свинья: ')
    number = int(input('Введите кол-во животных для покупки: '))
    costs = {'корова':100000, 'коза':9000, 'курица':2500, 'свинья':20000}
    cost = costs[animal_type]
    if self.money < number * cost:
      print('Деняк нет')
    else:
      self.money -= number * cost
      self.farm.add_animals(animal_type, number)
      print(f'Ура, {self.name} купил {number} {animal_type}!')

  def buy_seeds(self):   # покупаем семена
    seeds_type = input('Введите название семян, которые хотите купить - картошка, огурцы, помидоры, кукуруза, перец, горох, кабачки: ')
    number = int(input('Введите кол-во семян для покупки: '))
    costs = {'картошка':5, 'огурцы':1, 'помидоры':1, 'кукуруза':3, 'перец':3, 'горох':1, 'кабачки':4}  # цена за 1 шт/семечко, да...
    cost = costs[seeds_type]
    if self.money < number * cost:
      print('Деняк нет')
    else:
      self.money -= number * cost
      self.farm.add_seeds(seeds_type, number)
      print(f'Ура, {self.name} купил {number} {seeds_type}!')

  def sell_vegetables(self):     # продаем овощи
    vegetable_type = input('Введите название овоща, который хотите продать: картошка, огурцы, помидоры, кукуруза, перец, горох, кабачки: ')
    cost = int(input('По какой цене за шт. хотите продать? '))
    number = int(input('Сколько шт. хотите продать? '))
    if vegetable_type in self.farm.vegetable_fields.keys() and self.farm.vegetable_fields[vegetable_type] >= number:
      self.farm.vegetable_fields[vegetable_type] -= number
      self.money -= number * cost
      print(f'Ура, {self.name} продал {number} {vegetable_type} за {number * cost} руб!')
    elif vegetable_type in self.farm.vegetable_fields.keys() and self.farm.vegetable_fields[vegetable_type] < number:
      print(f'У {self.name} меньше овощей, чем он хочет продать.')
    elif vegetable_type not in self.farm.vegetable_fields.keys():
      print(f'У {self.name} нет таких овощей...')

  def sell_animals(self):     # продаем животных
    animal_type = input('Введите название животного, которое хотите продать: корова, коза, курица, свинья: ')
    cost = int(input('По какой цене за шт. хотите продать? '))
    number = int(input('Сколько шт. хотите продать? '))
    if animal_type in self.farm.my_animals.keys() and self.farm.my_animals[animal_type] >= number:
      self.farm.my_animals[animal_type] -= number
      self.money -= number * cost
      print(f'Ура, {self.name} продал {number} {animal_type} за {number * cost} руб!')
    elif animal_type in self.farm.my_animals.keys() and self.farm.my_animals[animal_type] < number:
      print(f'У {self.name} меньше животных данного вида, чем он хочет продать.')
    elif animal_type not in self.farm.my_animals.keys():
      print(f'У {self.name} нет таких животных...')

  def act_with_animal(self):  # что можно сделать с животным
    animal = input('С каким животным хотите взаимодействовать? ')
    if animal not in self.farm.my_animals.keys():
      print(f'У {self.name} нет такого животного!')
    else:
      action = input('Что вы хотите сделать с животным?')
      if action == 'подоить':
        self.farm.milk_animal()
      elif action == 'зарезать':
        self.farm.kill_animal()

  def sell_meat(self): # продаем мясо
    animal = input('Введите название животного, мясо которого хотите продать: корова, коза, курица, свинья: ')
    if animal not in self.farm.my_animals.keys():
      print(f'У {self.name} нет таких животных.')
    else:
      cost = int(input('По какой цене за кг. хотите продать? '))
      kilos = int(input('Сколько кг. хотите продать? '))
      if animal not in self.farm.meat.keys():
        print(f'У {self.name} нет мяса таких животных.')
      else:
        if self.farm.meat[animal] <  kilos:
          print(f'У {self.name} меньше мяса, чем он хочет продать.')
        else:
          self.farm.meat[animal] -= kilos
          self.money += kilos * cost
          print(f'Ура, {self.name} продал мясо за {kilos * cost} руб!')


  def sell_milk(self): # продаем молоко
    animal = input('Введите название животного, мясо которого хотите продать: корова, коза, курица, свинья: ')
    if animal not in self.farm.my_animals.keys():
      print(f'У {self.name} нет таких животных.')
    else:
      cost = int(input('По какой цене за литр хотите продать? '))
      liters = int(input('Сколько литров хотите продать? '))
      if animal not in self.farm.milk.keys():
        print(f'У {self.name} нет молока таких животных.')
      else:
        if self.farm.milk[animal] <  liters:
          print(f'У {self.name} меньше молока, чем он хочет продать.')
        else:
          self.farm.milk[animal] -= liters
          self.money += liters * cost
          print(f'Ура, {self.name} продал молоко за {liters * cost} руб!')




class Farm:
  def __init__(self):
    self.house = 1
    self.vegetable_fields = {}    # {овощи:кол-во}
    self.my_animals = {}  # {животные:кол-во}
    self.milk = {}   # {чье молоко:кол-во}
    self.meat = {}   # {чье мясо:кол-во}

  def add_animals(self, animal_type, number):
    if animal_type in self.my_animals.keys():
      self.my_animals[animal_type] += number
    else:
      self.my_animals[animal_type] = number

  def add_seeds(self, seeds_type, number):
    if seeds_type in self.vegetable_fields.keys():
      self.vegetable_fields[seeds_type] += number  # будем считать, что из каждого семечка, посаженного на поле, точно вырастет растение
    else:
      self.vegetable_fields[seeds_type] = number

  def milk_animal(self):  # для каждого животного свое кол-во молока
    animal = input('Какое животное хотите подоить - корова, коза: ')
    if animal == 'корова':
      liters = random.randint(8, 15)
      if animal in self.milk:
        self.milk[animal] += liters
      else:
        self.milk[animal] = liters
    if animal == 'коза':
      liters = random.randint(1, 3)
      if animal in self.milk:
        self.milk[animal] += liters
      else:
        self.milk[animal] = liters

  def kill_animal(self): # для каждого животного свой весовой диапазон
    animal = input('Какое животное хотите зарезать - корова, коза, курица, свинья: ')
    self.my_animals[animal] -= 1
    if animal == 'корова':
      kilos = random.randint(200, 400)
      if animal in self.meat.keys():
        self.meat[animal] += kilos
      else:
        self.meat[animal] = kilos
    if animal == 'свинья':
      kilos = random.randint(50, 150)
      if animal in self.meat.keys():
        self.meat[animal] += kilos
      else:
        self.meat[animal] = kilos
    if animal == 'курица':
      kilos = random.randint(1, 7)
      if animal in self.meat.keys():
        self.meat[animal] += kilos
      else:
        self.meat[animal] = kilos



class Game:
  def __init__(self, farmer):
    self.farmer = farmer

  def play_game(self):
    while True:
      print(f'Выберите, какое действие будет выполнять {self.farmer}. Введите соответствующую цифру:')
      print('1) купить семена', '2) купить животных', '3) продать овощи', '4) продать животных', '5) продать мясо', '6) продать молоко', '7) зарезать животное', '8) подоить животное', '9) показать характеристики', '10) вы вышли из игры', sep="\n")
      action = int(input('Цифра: '))
      if action == 10:
        print('Вы вышли из игры.')
        break
      if action == 9:
        print('Какие именно характеристики вы хотите посмотреть. Выберите соответствующую цифру:')
        print(f'1) сколько у {self.farmer} денег', f'2) сколько у {self.farmer} молока', f'3) сколько у {self.farmer} мяса', f'4) Сколько у {self.farmer} животных', f'5) Сколько у {self.farmer} овощей', sep="\n")
        char_action = int(input('Цифра: '))
        if char_action == 1:
          print(self.farmer.money)
        if char_action == 2:
          if self.farmer.farm.milk == {}:
            print('Молока нет.')
          else:
            print(self.farmer.farm.milk)
        if char_action == 3:
          if self.farmer.farm.meat == {}:
            print('Мяса нет.')
          else:
            print(self.farmer.farm.meat)
        if char_action == 4:
          if self.farmer.farm.my_animals == {}:
            print(f'У {self.farmer} пока нет животных.')
          else:
            print(self.farmer.farm.my_animals)
        if char_action == 5:
          if self.farmer.farm.vegetable_fields == {}:
            print(f'У {self.farmer} пока нет овощей.')
          else:
            print(self.farmer.farm.vegetable_fields)
      if action == 1:
        self.farmer.buy_seeds()
      if action == 2:
        self.farmer.buy_animal()
      if action == 3:
        self.farmer.sell_vegetables()
      if action == 4:
        self.farmer.sell_animals()
      if action == 5:
        self.farmer.sell_meat()
      if action == 6:
        self.farmer.sell_milk()
      if action == 7:
        self.farmer.act_with_animal()
      if action == 8:
        self.farmer.act_with_animal()
```
