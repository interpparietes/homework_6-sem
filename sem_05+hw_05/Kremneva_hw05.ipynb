#### Задача 1 (5 баллов).

Напишите игрушку Тамагочи (достаточно одного типа животного, но если пропишете классы еще для парочки, будет хорошо). Вам понадобятся модули random и sleep. У класса животного должны быть методы "спать", "играть" и "есть", а в атрибутах должны быть характеристики, например, "сытость" и "усталость". Лучше всего будет написать второй класс Game, в котором будет метод "играть", а экземпляр класса для выбранного животного будет храниться в его атрибутах. В процессе игры пользователь в цикле выбирает, что ему делать с животным (этот цикл может быть внутри метода "играть"). Воспользуйтесь ipywidgets для того, чтобы при запуске методов животного показывать, как проходит время и как пополняется или уменьшается какая-то из характеристик животного.

```python
import random
import time
from ipywidgets import IntProgress
from IPython.display import display
```

```python
class Tamagotchi:
  def __init__(self, name):
    self.name = name
    self.hunger = 50 # у меня hunger будет голод, а не сытость
    self.fatigue = 50

  def sleep(self):
    if self.fatigue <= 0:
      self.fatigue == 0
      print(f'{self.name} не хочет спать!')
    else:
      n = random.randint(10, 20)
      print(f'{self.name} спит.')

      f = IntProgress(value=self.fatigue, min = 0, max=100, description='Усталость:')
      display(f)
      for i in range(n):
        f.value -= 1
        time.sleep(.1)

      self.fatigue -= n

  def play(self): # во время игры и голод, и усталость будут увеличиваться
    if self.fatigue >= 100:
      self.fatigue == 100
      print(f'{self.name} слишком устал!')
    else:
      n = random.randint(10, 20)
      k = random.randint(10, 20)
      print(f'{self.name} играет.')

      w = IntProgress(value=self.hunger, min=0, max=100, description='Голод:', style={'bar_color': 'green'})
      display(w)
      for i in range(k):
        w.value += 1
        time.sleep(.1)

      f = IntProgress(value=self.fatigue, min = 0, max=100, description='Усталость:')
      display(f)
      for i in range(n):
        f.value += 1
        time.sleep(.1)

      self.fatigue += n
      self.hunger += k

  def eat(self):
    if self.hunger <= 0:
      self.hunger == 0
      print(f'{self.name} не хочет есть!')
    else:
      n = random.randint(10, 20)
      print(f'{self.name} ест.')

      w = IntProgress(value=self.hunger, min=0, max=100, description='Голод:', style={'bar_color': 'green'})
      display(w)
      for i in range(n):
        w.value -= 1
        time.sleep(.1)

      self.hunger -= n

  def show_charact(self):
    print(f'Голод: {self.hunger}')
    print(f'Усталость: {self.fatigue}')


class Game:
  def __init__(self, animal):
    self.animal = animal

  def play_game(self):
    while True:
      action= input('Выберите действие: спать, играть, есть, выйти, показать характеристики: ')
      if action == 'спать':
        self.animal.sleep()
      elif action == 'играть':
        self.animal.play()
      elif action == 'есть':
        self.animal.eat()
      elif action == 'показать характеристики':
        self.animal.show_charact()
      elif action == 'выйти':
        print('Вы вышли из игры.')
        break
      else:
        print('Неверное действие.')
```

```python
animal_name = input("Введите имя животного: ")
animal = Tamagotchi(animal_name)
game = Game(animal)
game.play_game()
```

Напишите функцию, которая будет отрисовывать на графике (столбцовая диаграмма подойдет) частоты для n самых частотных слов, при этом добавьте функции интерактивность и возможность выбирать n.

```python
import re
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets
from ipywidgets import interact, IntSlider
```

```python
text = []
with open('/content/turgenev.txt', 'r', encoding = 'utf8') as file:
  for line in file:
    text.extend([re.sub(r'[^\w\s]', '', word) for word in line.lower().split()])
# … « — эти товарищи убираются только регулярками... ни strip.punctuation, ни maketrans не работают, поэтому так

for i in range(len(text)):
  word = text[i]
  new_word = ""
  for k in word:
    if k.isdigit() or k in ('zyxwvutsrqponmlkjihgfedcbaéèçàäê'): # удаляю латынь и французский + числа
       new_word += ""
    else:
      new_word += k
  text[i] = new_word

for word in text: # удаляю пустые строки
  if word == '':
    text.remove(word)

statistics = Counter(text)
data = pd.DataFrame(sorted(statistics.items(), key=lambda item: item[1], reverse=True), columns=['word', 'frequency']) # делаем датафрейм из словаря


print(text)
print(statistics)
```

```python
def n_freq_word(df = ipywidgets.fixed(data), n = 5): # функция, рисующая график
  df = df.head(n)
  fig = plt.figure()
  ax = fig.add_subplot(111)
  bars = plt.bar(df['word'], df['frequency'], color='lightsalmon')
  plt.show()

interact(n_freq_word, n=IntSlider(min=1, max=len(data), step=1, value=1, description='Сколько слов хотите вывести? '))
```
