Поработаем с датасетом Titanic. Это очень известный датасет для машинного обучения, а нам он пригодится, чтобы попрактиковаться с pandas.

#### Задача 1.

Считайте датасет (можно скопировать код из лекции). Выведите первые пять строчек датасета. Посмотрите информацию о датасете. Во-первых, пустые ячейки всегда мешают, а во-вторых, в машинном обучении неудобно работать с текстовыми данными и все стараются превращать в цифры. Давайте попробуем почистить датасет с этой позиции, но разумно:

- Пропущенные ячейки возраста, наверное, можно заполнить средним арифметическим
- Пол можно заменить на чиселки - 1 для одного варианта и 0 для другого (не у всех пол известен, но там таких мало, можно заменить на тот же 0)
- Столбцы Ticket, Cabin и Embarked, скорее всего, содержат не очень интересную информацию: можно от них просто избавиться.
- Столбец с именами пока оставим как есть.

```python
import pandas as pd
```

```python
titanic = pd.read_csv('/content/titanic.csv', sep=';')
print(titanic.info()) # смотрим информацию о датасете
titanic.head() # выводим первые пять строчек датасета
```

```python
mean = float(round(titanic['Age'].mean())) # вычисляем средний возраст
titanic['Age'] = titanic['Age'].fillna('mean')  # заполняем пропущенные ячейки возраста средним арифметическим
titanic.info()
```

```python
titanic['Sex'] = titanic['Sex'].replace(['male', 'female', 'unknown'], [1, 0, 0]) # заменим пропуски и female на 0, male на 1
titanic.head()
```

```python
titanic.drop(['Ticket', 'Cabin', 'Embarked'], axis = 1 , inplace = True ) # удаяем эти три столбца
titanic.info()
```

#### Задача 2.

Сколько пассажиров выжило, а сколько - нет?

```python
print(f'Выжило: {titanic.Survived.sum()} чел., погибло: {len(titanic) - titanic.Survived.sum()} чел.')
```

#### Задача 3.

Какова доля выживших пассажиров из класса 3? А пассажиров из класса 1? (Доля - это отношение количества выживших пассажиров этого класса к общему количеству выживших пассажиров)

```python
print(f'Доля выживших пассажиров класса 3: {sum((titanic.Survived == 1) & (titanic.Pclass == 3)) / sum(titanic.Pclass == 3)}')
print(f'Доля выживших пассажиров класса 1: {sum((titanic.Survived == 1) & (titanic.Pclass == 1)) / sum(titanic.Pclass == 1)}')
```

#### Задача 4.

Какова доля выживших женщин из первого класса? А доля выживших мужчин из 3 класса?

```python
print(f'Доля выживших женщин из 1 класса: {sum((titanic.Survived == 1) & (titanic.Pclass == 1) & (titanic.Sex == 0)) / sum((titanic.Pclass == 1) & (titanic.Sex == 0))}')
print(f'Доля выживших мужчин из 3 класса: {sum((titanic.Survived == 1) & (titanic.Pclass == 3) & (titanic.Sex == 1)) / sum((titanic.Pclass == 3) & (titanic.Sex == 1))}')
```

#### Задача 5.

Давайте заведем отдельный столбец с фамилией пассажира. Можно заметить, что в ячейке имени фамилия всегда пишется первой и через точку с запятой. Вам поможет lambda и split.

```python
titanic['Surname'] = titanic['Name'].apply(lambda x: x.split(';')[0])
titanic.head()
```

#### Задача 6.

Сколько всего получилось семей? Сколько человек в каждой семье? Сколько семей, в которых больше трех человек?

```python
print(titanic['Surname'].nunique(), '\n') # всего семей(уникальная фамилия = семья)
print(titanic.groupby('Surname')['Surname'].count().head(), '\n') # человек в каждой семье
print((titanic.groupby('Surname')['Surname'].count() > 3).sum()) # кол-во семей, в которых > 3 человек
```

#### Задача 7.

Создайте столбец "IsChild", который равен 1, если возраст меньше 20, и 0 иначе.

```python
titanic['IsChild'] = titanic['Age'].apply(lambda x: 1 if x < 20 else 0) # памагите... откуда там str, если должно быть float
titanic.head()
```

#### Задача 8.

Какова доля семей, в которых минимальный возраст меньше 20 (семьи с детьми)? По отношению к общему количеству семей.

```python
print((titanic.groupby('Surname')['Age'].min() < 20).sum() / titanic['Surname'].nunique())
```
