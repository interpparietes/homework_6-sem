```python
from string import punctuation
from collections import Counter
from conllu import parse_incr
```

```python
class Conllu:
  def __init__(self, conllu: str):
    self.conllu = conllu
    self.lines = []

  def read_conllu(self): # прочитать файл
    data_file = open(self.conllu, "r", encoding="utf-8")
    for tokenlist in parse_incr(data_file):
      self.lines.append(tokenlist)
      print(tokenlist)

  def token(self, sentence_number, token_number): # вывести значения для конкретного токена из конкретного предложения
    token = self.lines[sentence_number][token_number]
    return token

  def poscount(self):
    count = []

    for token_list in self.lines:
      for token in token_list:
        for key in token.keys():
          if key == 'upos':
            count.append(token[key])

    count = Counter(count)

    print(f'POS{"":>13}%')

    word_sum = 0
    for value in count.values():
      word_sum += value

    for key, value in count.items():
      indent = 15 - len(key) - len(str(round((value/word_sum * 100))))
      print(f'{key}' + " " * indent + f'{round((value/word_sum * 100), 1)}')

  def unique_words(self):
    unique_words = []

    for token_list in self.lines:
      for token in token_list:
        for key in token.keys():
          if key == 'lemma' and token[key] not in unique_words:
            unique_words.append(token[key])

    print(unique_words)
    print(f'Количество уникальных слов: - {len(unique_words)}.')

  def lemmas(self):
    a = 0

    for token_list in self.lines:
      for token in token_list:
        if token['form'] == token['lemma']:
          a += 1

    print(f'Количество слов, которые по форме совпадают со своей леммой: {a}.')
```

```python
my_conllu = Conllu('/content/syntagrus_tiny.conllu')
my_conllu.read_conllu()
```

```python
my_conllu.poscount()
my_conllu.lemmas()
my_conllu.unique_words()
```
